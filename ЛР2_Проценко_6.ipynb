### 4. Проверка дубликатов

#### Проверка явных дубликатов

Далее используется метод duplicated, который считает дубликатами те строки, которые полностью повторяют уже встречавшиеся. Сначала выводится список всех дубликатов

```python
df[df.duplicated()]
```

Используется метод drop_duplicates для удаления строк-дублтикатов. Строки удаляются вместе с индексами, с помощью reset_index создаётся новый датафрейм, в котором не будет пропусков в индексах, drop=True чтобы не создавался новый столбец index.

```python
df = df.drop_duplicates().reset_index(drop=True)
```

#### Проверка неявных дубликатов

### 7. Матрица корреляции и тепловая карта 

```python
df.corr(numeric_only = True)
```

df.corr cчитает матрицу корреляций между всеми числовыми столбцами DataFrame. Параметр numeric_only=True означает, что учитываются только числовые колонкии возвращает таблицу, где показаны попарные корреляции между переменными

```python
df.corrwith(df['Income'], numeric_only = True)
```

df.corrwith cчитает корреляции всех числовых столбцов конкретно с колонкой 'Income'. Показывает, как каждая переменная связана с доходом

```python
df['Year_Birth'].corr(df['Income'])
```

Была подсчитана корреляция Пирсона между двумя конкретными колонками и показына прямая связь между годом рождения и доходом

В тепловой карте для представления значений данных используются разные оттенки цветов. Для создания тепловой карты использовался heatmap с параметрами: annot (отображение числовых значений коэффициентов), cmap (задает цветовую карту) , linewidth (устанавливает толщину линий, разделяющих ячейки)

```python
df_corr=df.corr(numeric_only = True)
sns.heatmap(df_corr,annot=True, cmap='BuPu', linewidths=.5).set_title('Тепловая карта корреляции')
```

Тепловая карта отлично подходит для визуализации корреляции, цветовая палитра делает визуализацию интуитивно понятной и позволяет быстро определить степень корреляции между переменными.

Задание 2. Использовать pandas и plot. По сводной таблице (pivot_table) -

Задание 3. Использовать matplotlib. Отфильтровать клиентов по year_birth
больше 1980. Построить круговую диаграмму, которая отображает процент
клиентов каждого семейного положения.

Сначала выполняется фильтрация датасета и остаются только клиенты родившиеся позже 1980 года. Затем с помощью метода value_counts() подсчитывается количество клиентов в каждой категории семейного положения.

```python
f_df = df[df['Year_Birth'] > 1980]
ms_counts = f_df['Marital_Status'].value_counts()
ms_counts
```

Для создания диаграммы импортирована библиотека Matplotlib. Устанавливается размер графика, и с помощью pie() на основе данных из ms_counts создается круговая диаграмма. Параметр labels использует индексы ms_counts в качестве меток сегментов, параметр autopct указывает как отображать проценты на сегментах, startangle устанавливает начальный угол для первого сегмента.
legend() добавляет легенду, параметр loc определяет местоположение легенды, параметр prop() устанавливает размер шрифта для легенды.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 8))
fig = plt.pie(ms_counts, labels=ms_counts.index, autopct='%1.1f%%', startangle=20)
plt.legend( loc='upper right', title ='Семейное положение', prop={'size': 8})
plt.title('Распределение клиентов по семейному положению')
```

Из данной диаграммы хорошо видно, какой процент клиентов имеет определенное семейное положение. Наибольшая часть клиентов женаты. На втором месте по количеству клиенты с семейным положением Single.

### 11. График типа boxplot по новой категории

Создана функциия categorize_income, которая принимает числовые значения Income, и возвращает категории на основе заданных условий. Затем функция применяется к столбцу "Income". В результате создается новый столбец "Income_Category", в которых каждая запись получает свою категорию в соответствии с условиями.

```python
def categorize_income(income):
    if income < 30000:
        return "Низкий доход"
    elif 30000 <= income < 80000:
        return "Средний доход"
    else: 
        return "Высокий доход"
df['Income_Category'] = df['Income'].apply(categorize_income)
df.head(20)
```

Для реализации диаграммы размаха для каждой категории дохода использовался boxplot

```python
sns.boxplot(data=df, x="Income", y="Income_Category")
```

Данная диаграмма размаха полезна для визуализации распределения данных и выявления выбросов. Черная диния - медиана (среднее значение дохода в категории). Ящик показывает разброс данных или степень изменчивости данных, в данном случае разброс относительно узкий у высокого и низкого дохода. Широкий разброс у среднего дохода. Круги - это выбросы, которые подразумевают, что есть данные, которые находятся далеко от основных данных и указывают на наличие нетипичных значений.
